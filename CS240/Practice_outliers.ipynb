# Loads libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loads Breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
df_breast_cancer = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
df_breast_cancer
```

```python
Q1 = df_breast_cancer.quantile(0.25)
Q3 = df_breast_cancer.quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = ((df_breast_cancer <= lower_bound) | (df_breast_cancer >= upper_bound)).any(axis=1)
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(data=df_breast_cancer)
plt.xticks(rotation=90)
plt.title('Boxplot of Breast Cancer Features with Outliers')
plt.show()
```
